# Importando `iris`

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data
```

# Particionando el conjunto

```python
X_train, X_test, y_train, y_test \
    = train_test_split(iris.data, iris.target, train_size=0.75)
```

```python
mensaje = f"""
Dimension de conjunto de datos
X_train: {X_train.shape}
y_train: {y_train.shape}
X_test:  {X_test.shape}
y_test:  {y_test.shape}
"""
print(mensaje)
```

# Version con `pandas`

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
X = iris.drop(columns=["Id", "Species"])
X
```

```python
# Con sklearn
# -----------------------------------------------
# from sklearn.preprocessing import LabelEncoder
# target = iris["Species"]
# label_encoder = LabelEncoder()
# target = label_encoder.fit_transform(target)

etiquetas = {
    "Iris-setosa":0,
    "Iris-versicolor":1,
    "Iris-virginica":2
}
y = iris["Species"].map(etiquetas)
y
```

```python
X_train, X_test, y_train, y_test \
    = train_test_split(X, y, train_size=0.75)
```

```python
mensaje = f"""
Dimension de conjunto de datos
X_train: {X_train.shape}
y_train: {y_train.shape}
X_test:  {X_test.shape}
y_test:  {y_test.shape}
"""
print(mensaje)
```
