# Métricas de evaluación de modelos

Cuando entrenamos modelos de ML es esencial tener maneras de evaluar su rendimiento. Para las tareas de clasificacion algunas de las métricas más comunes incluten la precisión (accuracy), la matriz de confunsión, la precisión (precision), la exhausitivdad (recall) y la puntuación F1.

## Accuracy

La precisión es la fracción de predicciones que el modelo acertó.

$$
\text{accuracy}=\frac{TP+TN}{TP+TN+FP+FN}
$$

donde
$$
\begin{aligned}
\text{TP} &= \text{True Positive} \\
\text{TN} &= \text{True Negative} \\
\text{FP} &= \text{False Positive} \\
\text{FN} &= \text{False Negative}
\end{aligned}
$$
o
$$
\text{accuracy} = \sum_{i=1}^nI(y_i = \hat y_i)
$$

## Matriz de confusión

$$
\begin{pmatrix}
  TP && TN\\
  FP && FN
\end{pmatrix}
$$

## F1 Score

El F1 Score es una medida que combina la precisión y la eshausividad. Es la media armonica de mabas medidas

## Error cuadrático medio (MSE)

El error cuadrático medio (MSE) es una métrica común para avaluar modelos de regresión. Mide le promedio de los cuadrados de los errores, es decir, la diferencia ciadrada media entre los valores estimados y los valores reales.
$$
\text{MSE} = \frac1n\sum_{i=1}^n(y_i-\hat y_i)^2
$$

# Libreria `sklearn`

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score,
                             recall_score, f1_score, classification_report)
```

```python
# 150 elementos en un grupo, 50 de otro grupo
X, y = make_moons(n_samples=(150,50))
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.grid()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
clf = MLPClassifier()
# clf = KNeighborsClassifier()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
y_pred, y_test
```

```python
print((y_pred == y_test).mean())
print(accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
```

```python
f1_score(y_test, y_pred)
```
